Гистограммы для оценки распределения GDP in 2023

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statsmodels.api as sm

df = pd.read_excel("my_full8.xlsx")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
pantone_287_rgb = (15/255, 45/255, 105/255)
pantone_2728_rgb = (55/255, 75/255, 155/255)

axes[0].hist(df['GDP in 2023'], bins=30, color=pantone_287_rgb, edgecolor='black', density=True)
axes[0].set_title('GDP in 2023 (Raw)')
axes[0].set_xlabel('GDP')
axes[0].set_ylabel('Density')
mean_raw = df['GDP in 2023'].mean()
std_raw = df['GDP in 2023'].std()
x_raw = np.linspace(df['GDP in 2023'].min(), df['GDP in 2023'].max(), 100)
axes[0].plot(x_raw, norm.pdf(x_raw, mean_raw, std_raw), 'k--', linewidth=2)

log_gdp = np.log(df['GDP in 2023'].replace(0, 1))
axes[1].hist(log_gdp, bins=30, color=pantone_2728_rgb , edgecolor='black', density=True)
axes[1].set_title('Log of GDP in 2023')
axes[1].set_xlabel('log(GDP)')
axes[1].set_ylabel('Density')
mean_log = log_gdp.mean()
std_log = log_gdp.std()
x_log = np.linspace(log_gdp.min(), log_gdp.max(), 100)
axes[1].plot(x_log, norm.pdf(x_log, mean_log, std_log), 'k--', linewidth=2)

plt.tight_layout()
plt.show()
```

модель для доли экспорта (с share of ex US)

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm

df = pd.read_excel("my_full8.xlsx")

df['log_gdp_country'] = np.log(df['GDP in 2023'])
df['log_val_export'] = np.log(df['val export in 2023'].replace(0, 1))
df['log_val_import'] = np.log(df['val import in 2023'].replace(0, 1))
df['log_ex_and_imp'] = np.log(df['ex_and_imp'].replace(0, 1))


numeric_columns = [
    'val export in 2023', '2023 Share export', 'val import in 2023', '2023 Share import',
    'USA sanctions', 'share of ex china', 'share of imp china', 'share of ex US',
    'share of imp US', 'US military', 'Kilometers', 'diplomatic status',
    'log_gdp_country', 'Population', 'area', 'ex_and_imp', 'v2x_libdem'
]
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')
df.columns = df.columns.str.replace(" ", "_")
df = df.rename(columns={'2023_Share_export': 'Share_export_2023'})

df["has_diplomatic_ties"] = df["diplomatic_status"].apply(lambda x: 1 if x > 0 else 0)

df["de_jure_status"] = df["diplomatic_status"].apply(lambda x: 1 if x == 2 else 0)


lm = sm.OLS.from_formula(
    'Share_export_2023 ~ USA_sanctions + share_of_imp_china + share_of_ex_US + share_of_imp_US + US_military + Kilometers + has_diplomatic_ties + de_jure_status + log_gdp_country + Population + area + v2x_libdem',
    df
)

model11 = lm.fit()

print("==== Модель для 2023 Share export ====")
print(model11.summary())

```

```python
df['residuals11'] = model11.resid

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(df['residuals11'], bins=5, color=pantone_2728_rgb, edgecolor='black')
plt.title('Гистограмма остатков для model11')
plt.xlabel('Остатки')
plt.ylabel('Частота')
```

```python
import seaborn as sns

fitted_vals = model11.fittedvalues
residuals = model11.resid

plt.figure(figsize=(10, 6))
sns.scatterplot(x=fitted_vals, y=residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Предсказанные значения (fitted values)")
plt.ylabel("Остатки (residuals)")
plt.title("Диаграмма рассеяния: остатки vs предсказанные значения")
plt.grid(True)
plt.show()
```
